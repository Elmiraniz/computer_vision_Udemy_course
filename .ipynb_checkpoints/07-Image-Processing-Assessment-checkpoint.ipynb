```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    shw = ax.imshow(img,cmap)
    plt.colorbar(shw)
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
img = cv2.imread('DATA/giraffes.jpg')
```

```python
img_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display_img(img_RGB)
```

**TASK:Apply a binary threshold onto the image.**

```python
img_gray = cv2.imread('DATA/giraffes.jpg', 0)
ret, img_thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
```

```python
display_img(img_thresh, cmap='gray')
```

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
display_img(img_hsv)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones((4,4), dtype=np.float32) / 10
```

```python
kernel
```

```python
img_blur = cv2.filter2D(img_RGB, -1, kernel)
```

```python
display_img(img_blur)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
sobelx = cv2.Sobel(img_gray, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
# the depth image (if given as short int CV_U, it is assumed to be the depth in millimeters (as done with the Microsoft Kinect), otherwise, if given as CV_32F or CV_64F, it is assumed in meters)
```

```python
display_img(sobelx, cmap='gray')
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
hist=np.zeros((3,256,1))
hist[0,:] = cv2.calcHist([img_RGB], channels=[0], mask=None, histSize=[256], ranges=[0,256])
hist[1,:] = cv2.calcHist([img_RGB], channels=[1], mask=None, histSize=[256], ranges=[0,256])
hist[2,:] = cv2.calcHist([img_RGB], channels=[2], mask=None, histSize=[256], ranges=[0,256])
```

```python
hist.shape
```

```python
colors = ['r', 'g', 'b']
fig = plt.figure()
for i,cl in enumerate(colors):
    plt.plot(hist[i,:], color=cl)
plt.title('Histogram for giraff')
```
